```python
x = np.linspace(-10,10,200)
```

```python
plt.plot(x, np.sin(x), 'r')
plt.plot(x, np.cos(x), 'g')
```

```python
plt.subplot(2,2,1)
plt.plot(x, np.sin(x), 'r')
plt.subplot(2,2,2)
plt.plot(x, np.cos(x), 'g')
plt.subplot(2,2,3)
plt.plot(x, x, 'g')
plt.subplot(2,2,4)
plt.plot(x, x**2, 'g')
```

```python
from sklearn.datasets import load_boston
boston = load_boston()
X = boston.data
Y = boston.target
```

```python
plt.figure(figsize=(8,10))
for i, feature in enumerate(boston.feature_names):
    plt.subplot(5,3,i+1)
    plt.scatter(X[:,i], Y, s=1)
    plt.ylabel('price')
    plt.xlabel(feature)
    plt.tight_layout()
```
