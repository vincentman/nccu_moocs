```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
from ipywidgets import interact
```

```python
x = np.linspace(0, 2*np.pi, 1000)
```

```python
def draw(n):
    y = np.sin(n*x)
    plt.plot(x, y, lw=5)
```

```python
draw(3)
```

```python
interact(draw, n=(1., 50.))
```

```python
def draw2(n):
    y = np.sin(n*x)
    plt.plot(x, y, lw=5)
    plt.show()
```

```python
interact(draw2, n=(1., 50.))
```
