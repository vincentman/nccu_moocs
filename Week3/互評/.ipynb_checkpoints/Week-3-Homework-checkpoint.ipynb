## ç¬¬ä¸€é¡Œ
é¡Œç›®ï¼šçµ¦å®šä¸€å€‹çŸ©é™£`m`ï¼Œè£¡é¢çš„å€¼éƒ½å¤§æ–¼0ã€‚è«‹å°‡çŸ©é™£mè£¡å¤§æ–¼ç­‰æ–¼1çš„æ•¸å­—éƒ½è®Šæˆ1ï¼Œå°æ–¼1çš„è®Šæˆ0ã€‚`m = np.array([[0,1,10],[2,6,7]])`ã€‚æç¤ºï¼šä½¿ç”¨å·¢ç‹€è¿´åœˆ(nested loop)åŠ åˆ¤æ–·å¼

---

* æœ¬é€±å‰›æ•™ `m[m>=1]` é€™ç¨® boolean indexingï¼Œé‚„èªªã€Œèƒ½ä¸ç”¨è¿´åœˆå°±ä¸ç”¨ã€ï¼Œæ­¤é¡Œå»æç¤ºç”¨å·¢ç‹€è¿´åœˆï¼Œå¾ˆæ€ª
* é¡Œç›®èªªã€Œ(çŸ©é™£m)è£¡é¢çš„å€¼éƒ½å¤§æ–¼0ã€ä½†ç¯„ä¾‹ä¸­ `m[0][0] == 0` å°±æ²’æœ‰å¤§æ–¼ 0

```python
# äº‚æ•¸ç”¢ç”Ÿä¸€å€‹è·Ÿé¡Œç›®å½¢ç‹€ç›¸åŒçš„ array (2, 3)
m = np.random.random_sample((2, 3))*2
m0 = m.copy()
m1 = m.copy()

# å·¢ç‹€è¿´åœˆè§£æ³•
print('åŸæœ¬çš„ m:\n%r' % (m,))
for i in range(m.shape[0]):
    for j in range(m.shape[1]):
        m[i][j] = 1 if m[i][j] >= 1 else 0
print('è™•ç†å¾Œ m:\n%r' % (m,))

# boolean indexing é…åˆæŒ‡å®šé‹ç®—è§£æ³•
i = m0 >= 1
m0[i] = 1
m0[np.logical_not(i)] = 0
print('m0ï¼Œç”¨ä¸åŒä½œæ³•ï¼Œè¦è·Ÿ m ç›¸ç­‰:\n%r' % (m0,))

# nditer ä¸€å±¤è¿´åœˆè§£æ³•
for x in np.nditer(m1, op_flags=['readwrite']):
    x[...] = 1 if x >= 1 else 0
print('m1ï¼Œç”¨ä¸åŒä½œæ³•ï¼Œè¦è·Ÿ m ç›¸ç­‰:\n%r' % (m1,))    

# é©—è­‰
# m è£¡åªæœ‰ 1 è·Ÿ 0
assert(np.logical_or(m == 1, m == 0).all())
# 3 å€‹è§£æ³•ç®—å‡ºä¾†ä¸€æ¨£
assert(np.array_equal(m, m0) and np.array_equal(m, m1))
```

## ç¬¬äºŒé¡Œ

```python
# 3ç­ã€æ¯ç­10äººã€5ç§‘
raw_score = np.random.randint(0, 100+1, (3, 10, 5))
weights = [.25, .25, .25, .15, .10] 

assert(np.sum(weights) == 1.0)
score = raw_score @ weights # `@` æ˜¯çŸ©é™£ä¹˜æ³• 
print('3ç­ã€æ¯ç­10äººçš„5ç§‘åŠ æ¬Šå¹³å‡æˆç¸¾ï¼š')
print(score)
```

## æŒ‘æˆ°é¡Œä¸€
è«‹å®šç¾©ä¸€å€‹å…¨ç­é«˜å¾®æˆç¸¾çš„å‡½å¼ï¼Œè¼¸å…¥æ˜¯ä¸€å€‹ (3, 30, 3) çš„ arrayï¼Œè¡¨ç¤º3å€‹ç­ç´šä¸­30ä½å­¸ç”Ÿçš„æˆç¸¾(å°è€ƒï¼ŒæœŸä¸­è€ƒï¼ŒæœŸæœ«è€ƒ)ã€‚ç›®æ¨™è¼¸å‡ºå‰‡æ˜¯ä¸€å€‹(3, 30, 1)çš„arrayï¼Œè¡¨ç¤ºæ‰€æœ‰å­¸ç”Ÿçš„æœŸæœ«æˆç¸¾ï¼ŒæœŸæœ«æˆç¸¾è¨ˆç®—æ–¹å¼å¦‚ä¸‹ï¼š
1. å°è€ƒ20%ï¼ŒæœŸä¸­è€ƒ35%ï¼ŒæœŸæœ«è€ƒ45%  
2. æœŸä¸­è€ƒ40%ï¼ŒæœŸæœ«è€ƒ60%ã€‚  
æ“‡å„ªä½œç‚ºæœŸæœ«æˆç¸¾
 
Hint: å¯ä»¥ç”¨é›™é‡forè¿´åœˆ åˆ†åˆ¥è·‘éç­ç´šè·Ÿäººã€‚å¯ä»¥å…ˆè¨­ä¸€å€‹arrayè£ç®—å‡ºä¾†çš„æˆç¸¾ã€‚è¨˜å¾—è¦ç”¨returnã€‚ è«‹åšå€‹æ¸¬è©¦ï¼Œæª¢é©—ä½ çš„çµæœï¼Œæ–¹ä¾¿åŒå­¸è©•åˆ†ã€‚

---

* ç¨‹å¼ä¸­ __`final_score()`__ å³ç‚ºé¡Œç›®è¦æ±‚çš„å‡½å¼
* æˆ‘æŠŠé¡Œç›®ã€Œè¼¸å‡ºå‰‡æ˜¯ä¸€å€‹(3, 30, 1)çš„arrayã€è§£é‡‹æˆè¼¸å‡ºè¦æ˜¯ shape ç‚º `(3, 30, 1)` çš„ _ä¸‰ç¶­_ arrayï¼Œä½†æœ‰äº›äººè¼¸å‡ºçš„å¯èƒ½æ˜¯ shape ç‚º `(3, 30)` çš„ _äºŒç¶­_ array
* æˆ‘ç•¶ç„¶èªç‚ºæˆ‘çš„è§£è®€æ¯”è¼ƒåˆç†ï¼Œå¦å‰‡é¡Œç›®å¯« `(3, 30)` å°±å¥½äº†ï¼Œä¸å¿…å¯« `(3, 30, 1)` ğŸ˜‰

```python
w0 = np.array([ .2,  .35, .45])
w1 = np.array([0.0,  .4,  .6 ])

def final_score(a):
    return np.maximum(a@w0, a@w1).reshape(a.shape[0], a.shape[1], 1)

# å°å‹æ¸¬è©¦
r0 = np.array([[[0, 100.0, 100.0], [100.0, 100.0, 0.0]]])
f0 = np.array([[[100.0], [55.0]]])
assert(np.array_equal(final_score(r0), f0))

r1 = np.array([[[0, 100.0, 100.0], [0, 0, 0]],
               [[0, 0, 0], [100.0, 100.0, 0]]])
f1 = np.array([[[100.0], [0.0]], [[0.0], [55.0]]])
assert(np.array_equal(final_score(r1), f1))
print('ç‚ºæ–¹ä¾¿ã€Œç›®æ¸¬ã€è©•åˆ†ï¼Œå…ˆå¸¶å…¥æ¯”è¼ƒå°é‡ (2, 2, 3) è³‡æ–™ã€‚')
print('åŸå§‹æˆç¸¾:\n%r' % (r1,))
print('åŠ æ¬Šæ³•0:\n%r' % ((r1@w0).reshape(r1.shape[0], r1.shape[1], 1),))
print('åŠ æ¬Šæ³•1:\n%r' % ((r1@w1).reshape(r1.shape[0], r1.shape[1], 1),))
print('æœŸæœ«æˆç¸¾(åŠ æ¬Šæ³•0ã€1ï¼Œæ“‡å„ª):\n%r' % (f1,))
print('é¡Œç›®è¦æ±‚çš„ (3, 30, 3) é™£åˆ—ï¼Œæ•¸å­—å¤šï¼Œå°±ç›´æ¥ç”¨ç¨‹å¼æ¯”è¼ƒï¼Œä¸å°å‡ºä¾†äº†ã€‚')

# äº‚æ•¸è³‡æ–™ï¼Œé¡Œç›®è¦æ±‚çš„ (3, 30, 3) é™£åˆ—
raw_score = np.random.randint(0, 100+1, (3, 30, 3))
s = final_score(raw_score)

# é©—è­‰ï¼š
# å½¢ç‹€å°
assert(s.shape == (3, 30, 1))
# æ¬Šé‡æ²’æ‰“éŒ¯
assert(np.sum(w0) == 1.0)
assert(np.sum(w1) == 1.0)
# å…©ç¨®ç®—æ³•æ“‡å„ª
assert((s >= (raw_score@w0).reshape(3, 30, 1)).all())
assert((s >= (raw_score@w1).reshape(3, 30, 1)).all())
# æœŸæœ«æˆç¸¾ä»‹æ–¼0åˆ°100
assert(np.logical_and(0 <= s, s <= 100).all())
```

## ä¸–ç´€é­”é¬¼æŒ‘æˆ°é¡Œï¼ˆåšå®Œå¯ä»¥ç›´æ¥å¾—åˆ°ä¸€ç™¾åˆ†ï¼‰
ç¾é£Ÿæ²™æ¼ çš„é¤é»æ¨è–¦ç³»çµ±
Susanæ˜¯å€‹æ•¸å­¸é«˜æ‰‹ï¼Œæ¯å¤©å°ˆå¿ƒç ”ç©¶é«˜æ·±æ•¸å­¸ç†è«–ï¼Œå› æ­¤æ²’æœ‰æ™‚é–“æƒ³æ¯ä¸€é¤è¦åƒä»€éº½ï¼Œ
è«‹ä¾ç…§ä»–çš„éœ€æ±‚å¹«ä»–åšå‡ºå„é¤çš„æ¨è–¦é¤é»ã€‚
 
çµ¦å®šä»¥ä¸‹list:(...)

è«‹éµå®ˆä»¥ä¸‹è¦å‰‡ï¼š
1. ä¾ç…§æ—©é¤-åˆé¤-æ™šé¤çš„é †åºæ±ºå®š
2. é¤é»å¿…é ˆç”¨äº‚æ•¸æ±ºå®š
3. é™¤äº†å®šç¾©ä»¥ä¸Šçš„listï¼Œä¸å¯ä»¥åœ¨å…¶ä»–åœ°æ–¹å‡ºç¾å­—ä¸²
 (ä¹Ÿå°±æ˜¯èªªä¸å¯ä»¥æ±ºå®šåˆé¤åƒ'å°å…¬å¯“'å¾Œï¼Œå¦å¤–å®šç¾©dinner = ['æ³¢æ³¢æ°æ°', 'æ¹¯é¥Œ', 'äº«çª©å’–å•¡',...]ï¼‰
4. ä¸€å¤©è£¡é¢åªå¯ä»¥æœ‰ä¸€é¤æ˜¯å±¬æ–¼high_price
5. åˆé¤è·Ÿæ™šé¤æˆ‘ä¸æƒ³åƒåŒä¸€å®¶
6. å¦‚æœç•¶å¤©ä¸‰é¤éƒ½æ²’æœ‰å±¬æ–¼high_priceçš„è©±ï¼Œå°±è¦åƒå®µå¤œ
 
### ç¯„ä¾‹è¼¸å‡º
Case 1:
æ—©é¤ï¼šå¤æ—©å‘³
åˆé¤ï¼šææ´›æ–¯
æ™šé¤ï¼šå¤§æ­£äº­
å®µå¤œï¼šé¹¹æ°´é›
Case 2:
æ—©é¤ï¼šå‚³é¦™é£¯ç³°
åˆé¤ï¼šå°å…¬å¯“
æ™šé¤ï¼šé¬¼åŒ 

---

* `meal_list_gen()` æœƒå‚³å›æ¨è–¦é¤é»ï¼Œå¦‚ï¼š`['å‚³é¦™é£¯ç³°', 'å‘·éºµé¨å£«', 'è´è¶æ¼¾', 'è±†èŠ±']`
* è¼¸å‡ºä¸­ï¼Œå³æ–¹æœ‰éŒ¢é£›èµ°äº†åœ–ç¤º(ğŸ’¸)ä»£è¡¨æ˜¯ high_price é¤å»³
* è¼¸å‡ºçš„ä¸‹ä¸€æ ¼ï¼Œæˆ‘ç”¨ç¨‹å¼å°‡ `meal_list_gen()` è·‘ `N` éï¼Œä¸¦é©—è­‰é¡Œç›®è¦æ±‚

```python
breakfast_list = ['å¤æ—©å‘³', 'å¾·åˆé¦™', 'å¸ƒæœ—å¥‡æ—©é¤', 'æ˜Ÿå·´å…‹', 'å°è²“å’ª', 'å‚³é¦™é£¯ç³°', 'ç±³å¡”', 'éº¥ç•¶å‹', 'æ‘©æ–¯', '7-11', 'å…¨å®¶']
lunch_and_dinner_list = ['å°å…¬å¯“', 'æ³¢æ³¢æ°æ°', 'æ¹¯é¥Œ', 'äº«çª©å’–å•¡', 'æ¹¯åŒ…', 'æ®ºäººé­”ç‹', 'å‘·éºµé¨å£«', 'ææ´›æ–¯', 'è´è¶æ¼¾', 'æ±äº¬å°åŸ', 'é‡‘é°­', 'æ¨‚å±±', 'Myéºµå±‹', 'ææ°é¤é…’é¤¨', 'é‡Œå…‹', 'æ°´ç…åŒ…', 'æ†©è³¢æ¨“', 'é«˜å¥éº—', 'é­¯æ™ºæ·±', 'æ•å¿ ', 'å¤§æ­£äº­', 'é˜¿é‡Œéƒ', 'é¬¼åŒ ', 'Juicy Bun', 'æ»‡å‘³å»šæˆ¿']
high_price_list = ['æ˜Ÿå·´å…‹', 'æ‘©æ–¯', 'å°å…¬å¯“', 'äº«çª©å’–å•¡', 'é‡‘é°­', 'ææ°é¤é…’é¤¨', 'é‡Œå…‹', 'Juicy Bun', 'æ»‡å‘³å»šæˆ¿']
snack_list = ['è±†èŠ±', 'æ”¿å¤§çƒ¤å ´', 'ä¸å¹«æ»·å‘³', 'é¹¹æ°´é›']

# é©—è­‰ high_price èˆ‡ snack ç„¡äº¤é›†
assert(not (set(high_price_list) & set(snack_list)))
# é©—è­‰ breakfast èˆ‡ lunch_and_dinner ç„¡äº¤é›†
assert(not (set(breakfast_list) & set(lunch_and_dinner_list)))

def meal_list_gen():
    '-> [æ—©é¤, åˆé¤, æ™šé¤ï¼Œ å®µå¤œ(å¯èƒ½æ˜¯ None)]'
    out = []
    for l in [breakfast_list, lunch_and_dinner_list, lunch_and_dinner_list]:
        t = np.random.choice(l)
        while (t in out) or ((t in high_price_list) and (set(out) & set(high_price_list))):
            t = np.random.choice(l)
        out.append(t)

    if not (set(out) & set(high_price_list)):
        out.append(np.random.choice(snack_list))
    else:
        out.append(None)
    return out

meals = meal_list_gen()
for n, m in zip(['æ—©é¤', 'åˆé¤', 'æ™šé¤', 'å®µå¤œ'], meals):
    if m is None:
        continue
    if m in high_price_list:
        end = '\tğŸ’¸\n'
    else:
        end = '\n'
    print('%sï¼š%s' % (n, m), end=end)
```

```python
# é©—è­‰
N = 1000
for i in range(N):
    meals = meal_list_gen()
    try:
        # meals[0] åœ¨æ—©é¤åˆ—è¡¨ï¼Œ 1ã€2 åœ¨åˆæ™šé¤åˆ—è¡¨
        assert(meals[0] in breakfast_list)
        assert(meals[1] in lunch_and_dinner_list)
        assert(meals[2] in lunch_and_dinner_list)
        # åˆã€æ™šé¤ä¸åŒ
        assert(meals[1] != meals[2])
        h = sum(1 for x in meals if x in high_price_list)
        # æœ€å¤šä¸€æ¬¡é«˜åƒ¹
        assert(h in {0, 1})
        # æ²’é«˜åƒ¹å°±åƒå®µå¤œï¼Œå¦å‰‡ä¸åƒ
        if not h:
            assert(meals[3] in snack_list)
        else:
            assert(meals[3] is None)
    except AssertionError as e:
        print(meals)
        raise
```
