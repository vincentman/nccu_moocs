```python
m = np.random.rand(3,10,5)
weights=[0.25,0.25,0.25,0.15,0.10]
```

```python
avg_cls = []
for cls in m:
    avg_stu = []
    for stu in cls:
        avg_stu.append(np.dot(stu, weights))
    avg_cls.append(np.sum(avg_stu)/len(avg_stu))
```

```python
print(avg_cls)
```
