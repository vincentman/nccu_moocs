```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
def select_good(a):
    weights1 = np.array([0.2, 0.35, 0.45])
    weights2 = np.array([0, 0.4, 0.6])
    grade1 = np.dot(a, weights1)
    grade2 = np.dot(a, weights2)
    return grade1 if grade1>=grade2 else grade2
```

```python
select_good(np.array([10, 20, 30]))
```

```python
def final_grade(a):
    result = np.zeros((3, 30, 1))
    for i in range(3):
        for j in range(30):
            result[i][j] = select_good(a[i][j])
    return result
```

```python
all_grade = np.zeros((3, 30, 3))
for i in range(3):
    for j in range(30):
        all_grade[i][j] = np.random.randint(0, 100, 3)
```

```python
final_grade(all_grade)
```
