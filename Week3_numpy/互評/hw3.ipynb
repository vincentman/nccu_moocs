# 第一題：
題目：給定一個矩陣m，裡面的值都大於0。
請將矩陣m裡大於等於1的數字都變成1，小於1的變成0。
 
m = np.array([[0,1,10],[2,6,7]])
 
提示：使用巢狀迴圈(nested loop)加判斷式

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

matrix = np.array([[0,1,10],[2,6,7]])

(m,n) = np.shape(matrix)

for i in range(m):
    for j in range(n):
        if(matrix[i][j]>=1):
            matrix[i][j] = 1
        elif(matrix[i][j]<1):
            matrix[i][j] = 0
            
print(matrix)
```

# 第二題：
有A,B,C三個班級，每班皆有10個學生，期中考結束後，老師希望計算各班級每位學生的加權平均成績。
 
請隨機產生3個班級的成績單,每班10人,共五科,權重為weights=[0.25,0.25,0.25,0.15,0.10]。
請算出班級加權平均成績，並以一個array輸出。

```python
grades = np.random.randint(0,100,(3,10,5))
weights=[0.25,0.25,0.25,0.15,0.10]
print(grades)

w_grades = grades * weights
avg_grades =np.sum(np.sum(w_grades,axis=2),axis=1)/10
print(avg_grades)
```

# 挑戰題一：
請定義一個全班高微成績的函式，輸入是一個(3, 30, 3)的array，  
表示3個班級中30位學生的成績(小考，期中考，期末考)  。

目標輸出則是一個(3, 30, 1)的array，表示所有學生的期末成績，
期末成績計算方式如下：

1. 小考20%，期中考35%，期末考45%  
2. 期中考40%，期末考60%。  
擇優作為期末成績
 
Hint
* 可以用雙重for迴圈 分別跑過班級跟人
* 可以先設一個array裝算出來的成績
* 記得要用return 
* 請做個測試，檢驗你的結果，方便同學評分。

```python
def grade_cal(list):
    w1 = [0.2, 0.35, 0.45]
    w2 = [0, 0.4, 0.6]

    
    g1 = list * w1
    g1 = np.sum(g1,axis=2)
    g2 = list * w2
    g2 = np.sum(g2,axis=2)
    m,n = np.shape(g1)
    result = np.zeros((m,n))
    
    for i in range(m):
        for j in range(n):
            result[i][j] = max(g1[i][j],g2[i][j])
    return result

grades = np.random.randint(0,100,(3,30,3))
print(grades)
grade_cal(grades)
```
